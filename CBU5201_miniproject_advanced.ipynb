```python
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input
from tensorflow.keras.preprocessing.image import img_to_array
from PIL import Image
import numpy as np
import os


def Load_Process(directory, image_size=(224, 224)):  # ResNet50 works better with 224x224 images
    image_files = [os.path.join(directory, file) for file in os.listdir(directory) if file.lower().endswith('.jpg')]
    preprocessed_images = []
    for file in image_files:
        image = Image.open(file).convert('RGB').resize(image_size)
        image_array = img_to_array(image)
        preprocessed_image = preprocess_input(image_array)
        preprocessed_images.append(preprocessed_image)
    return preprocessed_images, image_files

# Loading and preprocessing images
images, image_file_paths = Load_Process('./files')

# Load the pre-trained ResNet50 model without the top layer
model = ResNet50(include_top=False, input_shape=(224, 224, 3), pooling='avg')  # Added pooling='avg' for feature pooling

# Extract features using the model
def extract_features(images, model):
    features = model.predict(np.array(images), verbose=0)
    return features

# Extracting features from preprocessed images
extracted_features = extract_features(images, model)

# Using PCA for dimensionality reduction
from sklearn.decomposition import PCA
pca = PCA(n_components=50)
reduced_features = pca.fit_transform(extracted_features)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Clustering into two clusters
kmeans = KMeans(n_clusters=2, random_state=42)
clusters = kmeans.fit_predict(reduced_features)

# Plotting a few images from each cluster
def plot_images_from_cluster(cluster_idx, n_images=20):
    fig, axes = plt.subplots(1, n_images, figsize=(20, 4))
    fig.suptitle(f'Cluster {cluster_idx}', fontsize=16)
    selected_images = np.array(images)[clusters == cluster_idx][:n_images]
    for i, image in enumerate(selected_images):
        axes[i].imshow(image.astype('uint8'))
        axes[i].axis('off')
    plt.show()

# Display images from each cluster
plot_images_from_cluster(0)  # Images from cluster 0
plot_images_from_cluster(1)  # Images from cluster 1
```
